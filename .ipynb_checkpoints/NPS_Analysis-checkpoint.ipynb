```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

df = pd.read_csv("datasets/Student Feedback Surveys-Superview.csv",encoding= 'unicode_escape')
```

```python
df.rename(columns={'ï»¿ID': 'ID'}, inplace=True)
```

```python
df = df[df['Rating (Num)'] != '#ERROR!']
df['Rating (Num)'] = pd.to_numeric(df['Rating (Num)'], errors='ignore')
def get_nps(rating):
    if rating < 0 or rating > 10:
        raise ValueError('Number out of bounds')
    
    if 9 <= rating <= 10:
        return 'Promoter'
    
    elif 7 <= rating <= 8:
        return 'Passive'
    
    else:
        return 'Detractor'
ratings = []
for rating in df['Rating (Num)']:
    ratings.append(get_nps(rating))
```

```python
df['NPS_Type'] = ratings
```

```python
df
```

How many more promoters are there than detractors across our 2017 data?


```python
promoter_count = 0
detractor_count = 0
for nps in df['NPS_Type']:
    if nps is 'Promoter':
        promoter_count+=1
    elif nps is 'Detractor':
        detractor_count+=1
promo_detracto_diff = promoter_count-detractor_count       
```

```python
promo_detracto_diff
```

There are 639 more promoters

Which track boasts the best promoter-to-detractor ratio?


```python
pd.crosstab(df['Track'], df['NPS_Type'], normalize="index")
```

VR category boasts the best promoter to detractor ratio

Does student satisfaction vary by location?

```python
pd.crosstab(df['Location'], df['NPS_Type'], normalize="index")
```

Yes, the ratings by students do vary based on location. Atlanta promotes the most and Washington DC detracts the most

Now let's calculate total NPS

```python
def nps_calc(types):
    counts = types.value_counts()
    for each in ['Promoter', 'Passive', 'Detractor']:
        if each not in counts:
            counts[each] = 0
    difference_promoter_and_detractor = (counts['Promoter'] - counts['Detractor'])
    nps = difference_promoter_and_detractor / sum(counts) * 100
    return nps

nps_calc(df['NPS_Type'])
```
